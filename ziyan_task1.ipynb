```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('T1.csv',delimiter=',')
ages=df['age_difference']
#plotting histogram to visualize the distribution of age
ages.plot(kind='hist',bins=10,edgecolor='black')
#now cutomizing the histogram
plt.xlabel('Age Difference between character and actors')
plt.ylabel('Frequency')
plt.title('Age distribution')
plt.show()
```
